```
/* Run these program in dotnetfiddle */

/* Separate odd and even integers in separate arrays*/
using System;

public class Array
{
    public static void Main(string[] args)
    {
        Console.Write("Input the number of elements to be stored in the array: ");
        int n = int.Parse(Console.ReadLine());

        int[] arr = new int[n];
        Console.WriteLine("Input {0} elements in the array:", n);
        for (int i = 0; i < n; i++)
        {
            Console.Write("element - {0}: ", i);
            arr[i] = int.Parse(Console.ReadLine());
        }
        
        int[] evenArr = new int[n];
        int[] oddArr = new int[n];

        int evenCount = 0;
        int oddCount = 0;

        for (int i = 0; i < n; i++)
        {
            if (arr[i] % 2 == 0)
            {
                evenArr[evenCount] = arr[i];
                evenCount++;
            }
            else
            {
                oddArr[oddCount] = arr[i];
                oddCount++;
            }
        }

        Console.WriteLine("\n The Even numbers are:");
        for (int i = 0; i < evenCount; i++)
        {
            Console.Write(evenArr[i] + " ");
        }
        Console.WriteLine("\n The Odd numbers are:");
        for (int i = 0; i < oddCount; i++)
        {
            Console.Write(oddArr[i] + " ");
        }
        Console.WriteLine();
    }
}

/* Input the number of elements to be stored in the array: 5
Input 5 elements in the array:
element - 0: 25
element - 1: 47
element - 2: 42
element - 3: 56
element - 4: 32

The Even numbers are:
42 56 32
The Odd numbers are:
25 47 */
```
